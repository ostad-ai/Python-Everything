## Archimedean spiral
In *polar coordinates*, every point is represented by $(r,\theta)$
<br>A two-dimensional **spiral** may be defined in polar coordinates by $r=r(\theta)$
<br>To convert to *Cartesian coordinates*: 
 - $x=r(\theta)\,cos(\theta)$
 - $y=r(\theta)\,sin(\theta)$
 
**Example:** For **Archimedean spiral** we have:
- $r(\theta)=a\theta$, where $a$ is a constant real number

<br>The Python code is at: https://github.com/ostad-ai/Python-Everything

```python
# importing required modules
from matplotlib import pyplot as plt
import numpy as np
```

```python
a=2.5
def A_spiral(points=50):
    tetas=np.linspace(0,10*np.pi,points)
    r=a*tetas
    xs=r*np.cos(tetas)
    ys=r*np.sin(tetas)
    plt.plot(xs,ys,marker='*',ms=7,c=(.4,.5,0))
    plt.axis('off')
    plt.show()
A_spiral(points=70)
```
